```python
tmp = []# list
for each in raw_data['sex']:
    if each == 'female':#전체적으로 돌면서 여자일 경우 list에 0추가
        tmp.append(0)
    elif each == 'male': #남자일 경우 
        tmp.append(1)
    else: #아닐경우 추가x
        tmp.append(np.nan)
        
```

```python
raw_data['sex'] = tmp #위에서 배열에 [0,1,0,1]되어있는거 성별과 모두 교체
raw_data.head()
```

```python
raw_data['survived'] = raw_data['survived'].astype('float')
raw_data['pclass'] = raw_data['pclass'].astype('float')
raw_data['sex'] = raw_data['sex'].astype('float')
raw_data['sibsp'] = raw_data['sibsp'].astype('float')
raw_data['parch'] = raw_data['parch'].astype('float')
raw_data['fare'] = raw_data['fare'].astype('float')
raw_data.head() #정수 쓸 것 다 float화 
```

```python
raw_data = raw_data[raw_data['age'].notnull()]
raw_data = raw_data[raw_data['sibsp'].notnull()]
raw_data = raw_data[raw_data['parch'].notnull()]
raw_data = raw_data[raw_data['fare'].notnull()]#not null인 것들만 가져오기, null값 빼기 
raw_data.info()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(train_pre, #데이트프레임
                                   raw_data[['survived']], #맞추어야할 정답
                                   test_size = 0.1, #테스트데이터 10% 지정, 훈련데이터 90% 할당
                                   random_state = 13) #seed값, 섞기 , seed값지정하면 같게 셔플링 학습할떄 안정적인 데이터로   
                                    
                                     
```

```python
X_train = X_train.reset_index() #랜덤으로 섞은 인덱스 재정렬 
X_train = X_train.drop(['index'], axis=1) #인덱스열 삭제 axix=1 열단위로 작업. index열만 딱 제거 
# inplace = true를 써도 됨! axis=1 말고
X_test = X_test.reset_index()
X_test = X_test.drop(['index'], axis=1)
y_train = y_train.reset_index()
y_train = y_train.drop(['index'], axis=1)

y_test = y_test.reset_index()
y_test = y_test.drop(['index'], axis=1)
```

```python
from sklearn.tree import DecisionTreeClassifier
# 정확도가 100퍼센트가 되면안됨. 학습되지않은 상황에 대처해야하기 때문에.
#?? 100퍼센트가 이것도 while문 구조인가? 100%가 되면 멈추는것?
 #섞는 이유 학습하기 위해 . 정확도 1이면 학습종료! 실제로는 1이 없으므로 현실상황에 대응 x 지속적으로 틀린 것을 고쳐나가는 알고리즘생
#학습용 객체 생성
#파라메터를 분석가가 임의로 변경할 수 있음.
#파라메터를 분석가가 변경가능, 변경불가능
#분석가가 변경가능한 파라메터를 "하이퍼파라메터"
tree_clf = DecisionTreeClassifier(max_depth=3, random_state=13)
#공부시키기                      #트리의 최대 깊이, 값이 클 수록 복잡도 up
tree_clf.fit(X_train, y_train)   #3, 13의 경우는 테스트가 정확성이 높은데 왜 깊이가 깊어지면 공부데이터는 비례하는데 
                                 #테스트 데이터는 안그럴 때가 있는 것일까?
#train데이터를 가지고 공부한 점수
tree_clf.score(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score

y_pred = tree_clf.predict(X_test) #문제를 주고 풀어보라고 함. => 정답 return

#테스트데이터를 푼 답안과 해답을 비겨해서 정확도를 구해보자.
accuracy_score(y_test, y_pred)
```

```python
from sklearn.tree import export_graphviz

export_graphviz(
        tree_clf,
        out_file="titanic.dot",
        feature_names=['pclass', 'sex', 'age', 'sibsp', 'parch', 'fare'],
        class_names=['Unsurvived','Survived'],
        rounded=True,
        filled=True
    )
```
