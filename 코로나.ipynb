```python
timeDf = pd.read_csv('../covid/Time.csv', encoding = 'ISO-8859-1')
caseDf = pd.read_csv('../covid/Case.csv', encoding = 'ISO-8859-1')
pDf = pd.read_csv('../covid/PatientInfo.csv', encoding = 'ISO-8859-1')
```
